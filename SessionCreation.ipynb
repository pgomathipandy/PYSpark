```python
from pyspark.sql import SparkSession
spark= SparkSession.builder \
    .appName("EmployeeApp") \
    .getOrCreate()
print(spark)
```

```python
data = [
    (101, "Ram", "IT", 50000),
    (102, "Raghu", "HR", 60000),
    (103, "Sita", "Finance", 55000)
]

columns = ["EmployeeID", "Name", "Department", "Salary"]

df = spark.createDataFrame(data, columns)

df.show()
```

```python
from pyspark.sql.types import *
customer_schema = StructType([
    StructField("Customer_ID", StringType(), False),
    StructField("First_Name", StringType(), False),
    StructField("Last_Name", StringType(), False),
    StructField("Email", StringType(), False),
    StructField("Phone", StringType(), False),
    StructField("Country", StringType(), True),
    StructField("Join_Date", DateType (), True),
    StructField("Total_Spent", DoubleType(), True)
])
customer_df=spark.read.csv("/content/sample_data/customer.csv",header=True,schema=customer_schema)
customer_df.printSchema()
customer_df.show()
```

```python
customer_df.describe().show()
customer_df.summary().show()
```
