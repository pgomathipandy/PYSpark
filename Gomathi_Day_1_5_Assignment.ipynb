```python
from pyspark.sql import SparkSession

spark=SparkSession.builder.appName("StudentManagement").getOrCreate()

stud_schema=StructType([
    StructField('StudentID', IntegerType(), False),
    StructField('Name', StringType(), False),
    StructField('Age', IntegerType(), False),
    StructField('Course', StringType(), False),
    StructField('City', StringType(), True),
    StructField('Marks', IntegerType(), False)
    ])


stud_df=spark.read.csv("/content/students_07072026.csv",header=True,schema=stud_schema)
```

```python
stud_df.show(10)
stud_df.printSchema()
```

```python
stud_df.count()
```

```python
stud_df.select("Name","Marks").show(10)
```

```python
stud_df=stud_df.withColumn("Result",when((col("Marks")>=50),"Pass").otherwise("Fail"))
stud_df.show(10)
```

```python
stud_df=stud_df.withColumn("BonusMarks ",col("Marks")+5)
stud_df.show(10)
```

```python
employee_schema = StructType([
    StructField("EmpID", IntegerType(), False),
    StructField("Name", StringType(), False),
    StructField("Age", IntegerType(), False),
    StructField("Gender", StringType(), False),
    StructField("Department", StringType(), False),
    StructField("Salary", IntegerType(), False),
    StructField("City", StringType(), False)
])

employee_explicit = spark.read.csv("/content/employe_08072026.csv",header=True,schema=employee_schema)

employee_explicit.show(5)
employee_explicit.printSchema()
```

```python
employee_explicit=employee_explicit.withColumn("HRA",col("Salary")*0.15)

employee_explicit=employee_explicit.withColumn("DA ",col("Salary")*0.08)

employee_explicit=employee_explicit.withColumn("GrossSalary",col("DA ") + col("HRA") + col("Salary"))

employee_explicit.show()

```
